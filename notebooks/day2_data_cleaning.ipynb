```python
scheme_df = pd.read_csv("../data/processed/scheme_performance_clean.csv")

nav_df = pd.read_csv("../data/processed/nav_history_clean.csv")

txn_df = pd.read_csv("../data/processed/investor_transactions_clean.csv")

aum_df = pd.read_csv("../data/raw/03_aum_by_fund_house.csv")
```

```python
nav_df["date"] = pd.to_datetime(nav_df["date"])
txn_df["transaction_date"] = pd.to_datetime(txn_df["transaction_date"])
aum_df["date"] = pd.to_datetime(aum_df["date"])

all_dates = pd.concat([
    nav_df["date"],
    txn_df["transaction_date"],
    aum_df["date"]
]).drop_duplicates().sort_values()

dim_date = pd.DataFrame({"full_date": all_dates})

dim_date["day"] = dim_date["full_date"].dt.day
dim_date["month"] = dim_date["full_date"].dt.month
dim_date["year"] = dim_date["full_date"].dt.year
dim_date["quarter"] = dim_date["full_date"].dt.quarter
```

```python
type(dim_date)
```

```python
dim_date.to_sql(
    "dim_date",
    con=engine,
    if_exists="append",
    index=False
)

print("✅ dim_date loaded successfully")
```

```python
fund_dim = pd.read_sql("""
SELECT fund_id, amfi_code
FROM dim_fund
""", engine)

date_dim = pd.read_sql("""
SELECT date_id, full_date
FROM dim_date
""", engine)
```

# SQL Analysis Queries

```python
query1 = """
SELECT
    f.scheme_name,
    a.aum_crore
FROM fact_aum a
JOIN dim_fund f
ON a.fund_id = f.fund_id
ORDER BY a.aum_crore DESC
LIMIT 5;
"""

pd.read_sql(query1, engine)
```

```python
query2 = """
SELECT
    d.year,
    d.month,
    ROUND(AVG(n.nav_value), 2) AS avg_nav
FROM fact_nav n
JOIN dim_date d
ON n.date_id = d.date_id
GROUP BY d.year, d.month
ORDER BY d.year, d.month;
"""

pd.read_sql(query2, engine)
```

```python
query3 = """
SELECT
    df.scheme_name,
    fa.aum_crore
FROM fact_aum fa
JOIN dim_fund df
ON fa.fund_id = df.fund_id
ORDER BY fa.aum_crore DESC
LIMIT 5;
"""

pd.read_sql(query3, engine)
```

```python
query4 = """
SELECT
    df.scheme_name,
    fp.return_5yr_pct
FROM fact_performance fp
JOIN dim_fund df
ON fp.fund_id = df.fund_id
ORDER BY fp.return_5yr_pct DESC
LIMIT 5;
"""

pd.read_sql(query4, engine)
```

```python
query5 = """
SELECT
    df.category,
    ROUND(AVG(fp.expense_ratio_pct),2) AS avg_expense
FROM fact_performance fp
JOIN dim_fund df
ON fp.fund_id=df.fund_id
GROUP BY df.category
ORDER BY avg_expense;
"""

pd.read_sql(query5, engine)
```

```python
query6 = """
SELECT
    df.scheme_name,
    fp.sharpe_ratio
FROM fact_performance fp
JOIN dim_fund df
ON fp.fund_id=df.fund_id
ORDER BY fp.sharpe_ratio DESC
LIMIT 5;
"""

pd.read_sql(query6, engine)
```

```python
query7 = """
SELECT
    df.scheme_name,
    fp.expense_ratio_pct
FROM fact_performance fp
JOIN dim_fund df
ON fp.fund_id=df.fund_id
ORDER BY fp.expense_ratio_pct
LIMIT 5;
"""

pd.read_sql(query7, engine)
```

```python
query8 = """
SELECT
    df.scheme_name,
    fp.alpha
FROM fact_performance fp
JOIN dim_fund df
ON fp.fund_id = df.fund_id
ORDER BY fp.alpha DESC
LIMIT 5;
"""

pd.read_sql(query8, engine)
```

```python
query9 = """
SELECT
    df.scheme_name,
    fp.max_drawdown_pct
FROM fact_performance fp
JOIN dim_fund df
ON fp.fund_id = df.fund_id
ORDER BY fp.max_drawdown_pct DESC
LIMIT 5;
"""

pd.read_sql(query9, engine)
```

```python
query10 = """
SELECT
    df.category,
    ROUND(AVG(fp.return_5yr_pct),2) AS avg_return
FROM fact_performance fp
JOIN dim_fund df
ON fp.fund_id = df.fund_id
GROUP BY df.category
ORDER BY avg_return DESC;
"""

pd.read_sql(query10, engine)
```

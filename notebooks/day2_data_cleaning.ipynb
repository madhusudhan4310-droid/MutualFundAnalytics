```python
import pandas as pd
```

```python
nav_df = pd.read_csv("../data/raw/02_nav_history.csv")
```

```python
nav_df.head()
```

```python
nav_df.columns
```

```python
nav_df["date"] = pd.to_datetime(nav_df["date"])
```

```python
nav_df = nav_df.sort_values(["amfi_code", "date"])
```

```python
nav_df["nav"].isnull().sum()
```

```python
nav_df["nav"] = nav_df.groupby("amfi_code")["nav"].ffill()
```

```python
nav_df = nav_df.drop_duplicates()
```

```python
print("Number of rows:", len(nav_df))
```

```python
print((nav_df["nav"] > 0).all())
```

```python
nav_df.to_csv("../data/processed/nav_history_clean.csv", index=False)
```

```python
nav_df["nav"].isnull().sum()
```

# Cleaning investor_transactions.csv


```python
txn_df = pd.read_csv("../data/raw/08_investor_transactions.csv")
```

```python
txn_df.head()
```

```python
txn_df.columns
```

```python
txn_df["transaction_type"].unique()
```

```python
txn_df["transaction_type"] = (
    txn_df["transaction_type"]
    .astype(str)
    .str.strip()
    .str.title()
)
```

```python
txn_df["transaction_type"].unique()
```

```python
txn_df["transaction_type"] = txn_df["transaction_type"].replace({
    "Sip": "SIP",
    "Lumpsum": "Lumpsum",
    "Redemption": "Redemption"
})
```

```python
txn_df["transaction_date"] = pd.to_datetime(txn_df["transaction_date"])
```

```python
print((txn_df["amount_inr"] > 0).all())
```

```python
txn_df["kyc_status"].unique()
```

```python
valid = ["Verified", "Pending"]

print(txn_df["kyc_status"].isin(valid).all())
```

```python
txn_df.to_csv(
    "../data/processed/investor_transactions_clean.csv",
    index=False
)
```
